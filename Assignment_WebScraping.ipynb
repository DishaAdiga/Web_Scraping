```python
import requests
import html5lib
import bs4
```

```python
url = "https://www.keralaayurveda.biz/blog/stress-management-in-ayurveda"
r = requests.get(url)
htmlContent = r.text
print(htmlContent,'\n')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
title = soup.title
print(title)
```

```python
#Getting the introduction
intro = soup.find_all('p',class_='text-light')
for i in intro:
  print(i.get_text())
```

```python
#Getting all reference links
link = soup.find_all('a')
for i in link:
  print(i['href'],"\n")
```

```python
#Headings containing information
heading = soup.find_all('h3')
for i in heading:
  print(i.get_text(),"\n")
```

```python
#Get all li tags since information in the website is presented using html ul and li tags
for i in heading:
  ulList = soup.find_all('li')
  for li in ulList:
    print(li.get_text())
```
